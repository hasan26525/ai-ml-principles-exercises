```python
data_files = glob('sample_data/mnist*')
df = pd.read_csv(data_files[1],header=None)
print(df.info())
print(df.shape)

i = 3
img = df.iloc[i,1:].values.reshape(28,28)
img_title = df.iloc[i,0]
plt_img = plt.imshow(img)
plt.title(img_title,fontsize=14)
```

```python
pd.Series(img.flatten()).plot(kind = "hist",  bins = 50, title = "Distribution of Pixel's value")
plt.show()
```
